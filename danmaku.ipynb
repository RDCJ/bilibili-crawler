```python
keywords = ["虚拟up主","虚拟艺人","虚拟偶像","虚拟主播","vtuber","虚拟youtuber"]
```

```python
def dfsDict(dic):
    flag = False
    for value in dic.values():
        if type(value)==dic:
            flag = flag | dfsDict(value)
        else:
            strvalue = str(value).lower()
            for keyword in keywords:
                if strvalue.find(keyword) != -1:
                    flag = True
        if flag:
            return True
    return False 
```
